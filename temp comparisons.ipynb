```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
climate_change = pd.read_csv('Desktop/datacamp/matplotlib/climate_change.csv', index_col = 'date', parse_dates = True)
```

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'], color = 'b')
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change['relative_temp'], color = 'r')
plt.show()
```

```python
# to find out the rate of c02 emisions in 90's compared to early 2000's 
```

```python
seventies = climate_change['1970-01-01' : '1979-12-31']
```

```python
twenties = climate_change['2000-01-01' : '2010-12-31']
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(seventies.index,seventies['relative_temp'], color = 'r')
ax[1].plot(twenties.index, twenties['relative_temp'], color = 'b')
plt.show()
```

```python
# the visualization indicates that there's a significant increase in relative temp
```
